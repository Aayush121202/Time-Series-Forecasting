```python
import yfinance as yf
```

```python
ticker_symbol = 'TSLA'
```

```python
df = yf.download(ticker_symbol)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df['Open'].plot(figsize=(15,6))
```

# Simple Moving Average (Rolling Mean)

```python
df['Open:10 days rolling']= df['Open'].rolling(window=10,min_periods=1).mean()
```

```python
df[['Open','Open:10 days rolling']].plot(figsize=(15,6))
```

```python
df['Open:30 days rolling']= df['Open'].rolling(window=30,min_periods=1).mean()
```

```python
df['Open:50 days rolling']= df['Open'].rolling(window=50,min_periods=1).mean()
```

```python
df[['Open','Open:10 days rolling','Open:30 days rolling','Open:50 days rolling']].plot(xlim=['2023-01-01','2024-01-01'],figsize=(15,6))
```

# Cumulative Moving Average

```python
df['Open:Cumulative Moving Average']=df['Open'].expanding().mean().plot()
```

# Exponential Weight Moving Average (EWMA)

```python
df['Open:EWMA']=df['Open'].ewm(alpha=0.1, adjust=False).mean()
```

```python
df[['Open','Open:EWMA']].plot(xlim=['2023-01-01','2024-01-01'],figsize=(15,6))
```
